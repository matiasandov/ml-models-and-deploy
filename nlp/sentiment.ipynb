```python
import pandas as pd
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
```

```python
df = pd.read_csv("reviews.csv", sep=";", on_bad_lines="skip")
df
```

```python
df.dropna(subset=["Review.Text", "Rating"], inplace=True)
df
```

```python
reviews = df["Review.Text"]
reviews
```

```python
df.info()
```

```python
ratings = df["Rating"]

ratings = ratings.apply(lambda x: 'pos' if x > 3 else('neg' if x < 3 else 'neu'))
        
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

ratings = le.fit_transform(ratings)
from tensorflow.keras.utils import to_categorical
ratings = to_categorical(ratings)
```

```python
import numpy as np
np.unique(ratings, return_counts=True)
```

```python
import re
corpus = []
#ya remueve stopwords
stemm = SnowballStemmer(language="english")
for i in reviews:
    words = i.lower()
    words = re.sub("[^a-zA-Z]", " ", words)
    words = words.split()
    words = [stemm.stem(word) for word in words]
    words = " ".join(words)
    corpus.append(words)
    

    
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer()
x = vect.fit_transform(corpus)
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, ratings, test_size=0.2)
```

```python
import tensorflow
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization
from keras.callbacks import EarlyStopping

early = EarlyStopping(
    patience=8,
    monitor="recall",
    mode="max",
    restore_best_weights=True
)

model = Sequential()
model.add(Dense(1, activation="relu"))
model.add(BatchNormalization())
model.add(Dense(100, activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.1))
model.add(Dense(700, activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(700, activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(Dense(400, activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.1))
model.add(Dense(100, activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.1))
model.add(Dense(ratings.shape[1], activation="softmax"))

model.compile(loss= "categorical_crossentropy", optimizer="adam", metrics=["recall"])
model.fit(xtrain,ytrain, epochs=100, callbacks=[early], validation_split=0.1)
```

```python
print(model.evaluate(xtest,ytest))
```

```python
predictions = model.predict(xtest)
predicted_classes = np.argmax(predictions, axis=1)
true_classes = np.argmax(ytest, axis=1)

# Compare the predicted and true classes
for i in range(10):
    print(f'Predicted: {predicted_classes[i]}, True: {true_classes[i]}')
```
