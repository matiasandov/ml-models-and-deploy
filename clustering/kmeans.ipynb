```python
import pandas as pd
import sklearn
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("clustering.csv")
df
```

```python
df.info()
```

```python
df = df.drop(["Unnamed: 0"], axis=1)
df
```

```python
sc = StandardScaler()
x = sc.fit_transform(df)
x
```

metodo del codo

```python
from sklearn.cluster import KMeans
scores = []

for i in range(1,16):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(x)
    #inertia: distancias al cuadrado entre cada centroide
    scores.append(kmeans.inertia_)
    
import matplotlib.pyplot as plt
plt.plot( range(1,16), scores, "bx-" )
plt.xlabel("N clusters")
plt.ylabel("Distancias")
plt.show()
```

# Graph silhouette score
Evaluates how unique an object is to its own cluster compared to other clusters

```python
from sklearn.metrics import silhouette_score
silscores = []

for i in range(2,16):
    kmeans = KMeans(n_clusters=i, random_state=42)
    labels = kmeans.fit_predict(x)
    silscores.append(silhouette_score(x,labels))

plt.plot( range(2,16), silscores, "bx-")
plt.xlabel("N clusters")
plt.ylabel("Sil scores")
plt.show()
    
```

```python
model = KMeans(n_clusters=6, random_state=42)
```

```python
df["cluster"] = model.fit_predict(x)
df.info()
```

```python
df["cluster"].value_counts()
```
