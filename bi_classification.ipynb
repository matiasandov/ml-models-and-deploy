# Preprocessing 

```python
from  sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
for col in df.select_dtypes("object").columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    
df
```

```python
Y = df["Churn"]
X = df.drop(["customerID", "Churn"], axis=1)
```

```python
from sklearn.model_selection import train_test_split
#primero todas las XX
X_train, X_test, y_train, y_test = train_test_split(X,Y, train_size= 0.8, random_state=42)
```

```python
X_train
```

```python
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
selector = SelectKBest(score_func=f_classif, k=16)

# Fit the selector to the training data
selector.fit(X_train, y_train)

# Transform the data to select the top k features
X_train = selector.transform(X_train)
X_test = selector.transform(X_test)

# Get the selected feature indices and scores
selected_indices = selector.get_support(indices=True)
selected_scores = selector.scores_

print(selected_indices)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

params = {
    #coef for regularization - unidades 
    "C": [0.1, 1, 5, 10],
    #coef for kernel function - 
    "gamma": [1, 0.1 , 0.01, 0.05, 0.005, 0.001],
    "kernel": ["rbf", "linear", "sigmoid"]
}

grid = GridSearchCV(SVC(), scoring="recall", param_grid=params, cv=4, refit=True)

grid.fit(X_train, y_train)

y_pred = grid.best_estimator_.predict(X_test)
y_pred
```

```python
grid.best_estimator_
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB

params = {
    #coef for regularization - unidades 
    "var_smoothing": [1e-8, 1e-9, 1e-10],
    
}

grid = GridSearchCV(GaussianNB(), param_grid=params, scoring="recall",cv=4, refit=True)

grid.fit(X_train, y_train)

y_pred = grid.best_estimator_.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

params = {
    #coef for regularization - unidades 
    "C": [1.0, 0.5 , 0.1],
   
    
}

grid = GridSearchCV(LogisticRegression(), param_grid=params, scoring="recall", cv=4, refit=True)

grid.fit(X_train, y_train)

y_pred = grid.best_estimator_.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
sv = SVC(C=5, gamma=0.1, kernel="sigmoid")
sv.fit(X_train, y_train)
y_pred = sv.predict(X_test)
print(classification_report(y_test, y_pred))
```
